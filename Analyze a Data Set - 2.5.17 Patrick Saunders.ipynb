```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
data = pd.read_csv(os.getcwd() + '/boston_housing_data.csv', 
                   names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX',
                            'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 
                            'B', 'LSTAT', 'MEDV'],
                   delim_whitespace = True)
data
```

```python
print("median house value in $1000s accounting for all suburbs")
print(data.MEDV.median())
print("mean house value in $1000s accounting for all suburbs")
print(data.MEDV.mean())
```

```python
print("median value of suburb with the newest houses")
data[data.AGE == data.AGE.min()].MEDV
```

```python
plt.scatter(data.CRIM, data.PTRATIO)
plt.xlabel("crime rate")
plt.ylabel("pupil-teacher ratio")
plt.title("Impact of class size on crime rate")
plt.show()
```

There doesn't appear to be a strong relationship between crime rate atnd pupil-teacher ratio. The odd shape of the data makes me suspicious of the data quality.

```python
plt.scatter(data.B, data.DIS)
plt.xlabel("Proportion of black citezens")
plt.ylabel("distance to employment centers")
plt.title("Relationship between commute distance and race")
plt.show()
```

There's a pretty clear pattern of black citizens living far from employment centers here.

```python
plt.scatter(data.MEDV, data.NOX)
plt.xlabel("Median home value")
plt.ylabel("Nitric Oxide concentration")
plt.title("effect of nitric oxide on home value")
plt.show()
```

High nitric oxide concentrations tend to reduce home value, but they are clearly not the only factor.

```python
plt.scatter(data.INDUS, data.CRIM)
plt.xlabel("proprotion of non-retail buisness")
plt.ylabel("crime rate")
plt.title("relationship between non-retail business and crime")
plt.show()
```

I wanted to know how crime rate and industry were related.
There's a spike in crime rate on the higher end of industiral proportion, but not at the highest end. Maybe it's where there's mixed industrial and residential zoning?
