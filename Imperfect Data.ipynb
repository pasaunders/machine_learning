The general theme is, which streets are likely to be less crowded in Seattle
1) Which street has the highest recorded traffic
2) Which street has the lowest recorded traffic
3 & 4) Do downtown streets or non-downtown streets have a higher average traffic level
5) Which downtown street had the lowest traffic in the most recent year.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
data = pd.read_csv(os.getcwd() + '/Traffic_Flow_Counts.csv', delimiter=",",
                   names = ['downtown', 'stname', 'year', 'aawdt'])
data
```

Cleaning any incomplete data:

```python
data = data.fillna(data.mode())
```

```python
print("Highest recorded traffic")
print(data[data.aawdt == data.aawdt.max()].stname)
```

```python
print("Lowest recorded traffic")
print(data[data.aawdt == data.aawdt.min()].stname)
```

```python
print("average downtown traffic rating")
print(sum(data[data.downtown == True].aawdt / len(data[data.downtown == True])))
```

```python
print("average non-downtown traffic rating")
print(sum(data[data.downtown == False].aawdt / len(data[data.downtown == False])))
```

```python
print("Lowest recorded downtown traffic")
dataset = data[(data.downtown == True) & (data.year == data.year.max())]
print(dataset[dataset.aawdt == dataset.aawdt.min()].stname)
```
